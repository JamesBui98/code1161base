Seal Pup Count in Seal Bay from 1985-2015

Introduction

  Seal Bay, located in South Australia on the south coast of Kangaroo Island, is the third largest colony of Australia's seal population.
  Post 1945, Seals were being used as shark bait as well as oil and leather.
 Thus, Seal Bay is a conservation reserve to protect these endangered species 

However...

In recent findings, the Sea lion population is declining from the data shown which was recorded since 1985-2015

```python
pupcount_data.head(22)
```

```python
pupcount_data["Pup count"].plot(kind="bar")
plt.title("Pup Count from 1985-2015")
plt.xlabel("Pup Count", fontsize=20)
plt.ylabel("Year (1985-2015)", fontsize=20)
plt.grid(True)
plt.show()
```
